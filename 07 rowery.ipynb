### Najpierw ściągnij plik http://prokulski.science/temp/test/veturilo.zip i wypakuj do folderu "rowery"

### Co tu się będzie działo?

1. wczytamy wszystkie JSONy po kolei
2. zrobimy z nich wielki data frame
3. i zapiszemy do np. CSV

```python
import pandas as pd
import json
from pathlib import Path
```

```python
# przygotowanie listy wszystkich plików

folder_glowny = Path("rowery/veturilo/2025-10-20")
```

```python
def przetworz_plik_json(plik):
    # wczytujemy zawartość pliku
    dane_json = plik.read_text(encoding='utf-8')

    # parsujemy zawartość jako json
    dane = json.loads(dane_json)

    # dla każdej stacji z danych z pliku...
    dane_stacje = []
    for station in dane['network']['stations']:
        # usun klucz "extra"
        station.pop('extra')
        # dodaj do listy rekordów
        dane_stacje.append(station)

    # z całej listy zrób data frame
    dane_stacje_df = pd.DataFrame(dane_stacje)

    # oddaj poza funkcję
    return dane_stacje_df
```

```python
lista_df = []
for plik in folder_glowny.glob("*.json"): # wyszukaj wszytkie pliki z rozszerzeniem json
    temp_df = przetworz_plik_json(plik) # przetwórz każdy z nich
    lista_df.append(temp_df)
```

```python
def zmien_date(s):
    dt = pd.to_datetime(s[:26])
    return dt
```

```python
rowery_df['datetime'] = rowery_df['timestamp'].apply(lambda s: zmien_date(s))
```
