```python
import pandas as pd
```

```python
nazwa_pliku = 'pivot_z_excela.xlsx'
nazwa_arkusza = 'dane'
```

```python
df = pd.read_excel(nazwa_pliku, sheet_name=nazwa_arkusza)
df
```

```python
# kombajn na wszystko na raz
pd.pivot_table(df, values=['Wartosc', 'Cena'], columns='Sprzedawca', aggfunc=['sum', 'mean', 'size'])
```

```python
agg_1_df = pd.pivot_table(df, values='Wartosc', columns='Sprzedawca', aggfunc='sum')
agg_1_df
```

```python
agg_2_df = pd.pivot_table(df, values='Cena', columns='Sprzedawca', aggfunc='mean')
agg_2_df
```

```python
agg_3_df = pd.pivot_table(df, values='Ilosc', columns='Sprzedawca', aggfunc='sum')
agg_3_df
```

```python
wynik_df = pd.concat( [agg_1_df, agg_2_df, agg_3_df], axis=0 )
wynik_df
```

```python
wynik_df['Suma'] = wynik_df['Anna'] + wynik_df['Jan'] + wynik_df['Marek']
wynik_df
```
