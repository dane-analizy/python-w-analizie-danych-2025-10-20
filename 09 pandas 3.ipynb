```python
# zmieniamy nazwy wszystkich kolumn cecha_* na wartosc_*
for n, c in enumerate(df.columns, start=1):
    if c.startswith('cecha_'):
        df.rename(columns={c:f'wartosc_{n}'}, inplace=True)

df
```

# grupowanie

```python
data = {
    'kategoria': ['A', 'A', 'B', 'B', 'C', 'C', 'D'],
    'cena': [10, 15, 10, 20, 10, 30, 45],
    'ilosc': [1, 2, 1, 3, 1, 1, 0]
}
df = pd.DataFrame(data)

df
```

```python
df['wartosc'] = df['cena'] * df['ilosc']
df['dzielenie'] = df['cena'] / df['ilosc']
df
```

```python
df.loc[6, 'dzielenie'] == np.nan
```

```python
grupy = df.groupby('kategoria')
grupy
```

```python
for g_name, g_data in grupy:
    print(g_name)
    print(g_data)
    print()
    g_data.to_excel(f"raport_po_klientach/raport_{g_name}.xlsx", index=False)
```

```python
df.groupby('kategoria').size() # tu mamy indeks = nazwa kategorii z grupowania
```

```python
df = pd.DataFrame({
    "region": np.random.choice(["Północ", "Południe", "Wschód", "Zachód"], size=20),
    "produkt": np.random.choice(["A", "B", "C"], size=20),
    "sprzedawca": np.random.choice(["Jan", "Anna", "Tomek", "Ewa"], size=20),
    "rok": np.random.choice([2023, 2024], size=20),
    "sprzedaz": np.random.randint(100, 1000, size=20),
    "marza": np.random.uniform(0.05, 0.3, size=20).round(2)
})

df
```

```python
agg_df = df.groupby( ['region', 'produkt'], as_index=False).aggregate( {
  'sprzedaz' : 'mean',
    'marza': ['min', 'max', 'mean', 'median']
} )
agg_df
```
